```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
peaks = [0.85,0.8,0.75,0.66,0.5]
s1 = [0.67,0.66,0.61,0.55,0.53]
s2 = [0.80,0.80,0.80,0.79,0.78]
s3 = [0.66,0.73,0.65,0.59,0.6]
plt.plot(peaks, s1, color='red', marker='o',linewidth=3.0, label='GFT')
plt.plot(peaks, s2, color='olive', marker='o',linewidth=3.0,label='Graph-Wavelet')
plt.plot(peaks, s3, color='blue', marker='o',linewidth=3.0,label='CNN')
plt.ylim(ymin=0,ymax=1)
plt.legend(frameon=False)
plt.xlabel('Fraction of Training Data',size=20)
plt.ylabel('Accuracy of Model',size=20)
plt.title('Accuracy-Alpha/Beta Classification',size=15)
# plt.savefig("nearest_gene.png")
```
